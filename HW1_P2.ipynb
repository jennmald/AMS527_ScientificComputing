This notebook completes homework 1 problem 2. It will generate the first n terms of the sequence $x_{k+1}=111-(1130-3000/x_{k-1})/x_k$ given $x_1$ and $x_2$. Similar to problem 1, numpy float32 and float64 are used to complete the problem in both single and double precision. Code can be run or plots and analysis are below.

```python
"""
Author : Jennefer Maldonado
Date   : 2/9/2021
Purpose: Generate first n terms of a sequence in double
         and single precision
"""

import matplotlib.pyplot as plt
import numpy as np
import math

# single precision
def single_seq(n):
    x = [np.float32(11.0/2.0), np.float32(61.0/11.0)]
    for k in range(1,n):
        x_new = np.float32(111.)-(np.float32(1130.)- (np.float32(3000.)/x[k-1]))/x[k]
        x.append(x_new)
    return x


# double precision
def double_seq(n):
    x = [np.float64(11.0/2.0), np.float64(61.0/11.0)]
    for k in range(1,n):
        x_new = np.float64(111.)-(np.float64(1130.)- np.float64(3000.)/x[k-1])/x[k]
        x.append(x_new)
    return x

ns = 10
sing = single_seq(ns)

nd = 20
doub = double_seq(nd)

print(doub)
plt.plot(doub)
plt.title("Double Precision")
plt.ylabel("")
plt.xlabel("")
plt.show()

print(sing)
plt.plot(sing)
plt.title("Single Precision")
plt.ylabel("")
plt.xlabel("")
plt.show()
```

### Analysis 
Both graphs for the double and single precision show the sequence approach 6 but reach a point where it peaks above six. The double precision seems to converge to six then converge to 100, similar results occur in single precision as well. The discrempancy happens I believe when the division becomes too small. Thus, causing the weird change in the behavior of the graph compared to the theoretical behavior.
