```python
"""
Author : Jennefer Maldonado
Date   : 2/10/2021
Purpose: Compute quadratic function
"""


import math

def alternative(a,b,c):
    num = 2*c
    plus_denom = -1*b + math.sqrt(b**2 - (4*a*c))
    neg_denom = -1*b - math.sqrt(b**2 - (4*a*c))
    root1 = 0
    root2 = 0
    if plus_denom == 0:
        root1 = None
        root2 = num/neg_denom
    elif neg_denom == 0:
        root1 = num/plus_denom
        root2 = None
    return [root1, root2]


def quadratic(a,b,c):
    root1 = 0
    root2 = 0 
    try:
        plus_num = -1*b + math.sqrt(b**2 - (4*a*c))
        neg_num = -1*b - math.sqrt(b**2 - (4*a*c))
        denom = 2*a
        if denom != 0:
            root1 = plus_num/denom
            root2 = neg_num/denom
        else:
            print('Alternative Formula used')
            new_roots = alternative(a,b,c)
            root1 = new_roots[0]
            root2 = new_roots[1]
            
    except:
        print('Caught overflow error, scaling coefficents')
        a = str(a)
        b = str(b)
        c = str(c)
        aexp = a.split('e')
        bexp = b.split('e')
        cexp = c.split('e')
        
        ## check if they have exponent then drop exponent
        if (aexp[1] == bexp[1]) and (bexp[1] == cexp[1]):
            a = int(aexp[0])
            b = int(bexp[0])
            c = int(cexp[0])
                
        plus_num = -1*b + math.sqrt(b**2 - (4*a*c))
        neg_num = -1*b - math.sqrt(b**2 - (4*a*c))
        denom = 2*a
        if denom != 0:
            root1 = plus_num/denom
            root2 = neg_num/denom
    return [root1, root2]


def main():
    a = [6, 6e154, 0, 1, 1, 0]
    b = [5, 5e154, 1, -1e5, -4, -1]
    c = [-4, -4e154, 1, 1, 3.999999, 1]
    
    for i in range(0, len(a)):
        print('Coefficients: '+str(a[i])+', '+str(b[i])+', '+str(c[i]))
        roots = quadratic(a[i], b[i], c[i])
        print('root 1:' + str(roots[0]))
        print('root 2:' + str(roots[1]))
        print()
    

if __name__ == '__main__':
    main()    
```

## Analysis
I used two different methods for both versions of the formula. If the method for the original formula catches an exception there are two cases: overflow or divide by zero. The program scales the coefficients if overflow is the problem or switches formulas to avoid the divide by zero error. If at any point the numerator is zero in the formula, the default root is set to None type. The code produces similar results to the given solution table on the homework pdf, but has some more digits of precision.
