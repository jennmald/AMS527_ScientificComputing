```python
# Jennefer Maldonado
# Homework 6 Problem 6
# Solve the heat equation using an ODE solver

## import statements
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def ode_solver(N, M, xmin, xmax, tmin, tmax):
    # delta x 
    dx = float(xmax - xmin) / N
    # delta t
    dt = float(tmax - tmin) / M
    
    # define x values
    x = np.linspace(xmin, xmax, N)
    # define t values
    t = np.linspace(tmin, tmax, M)
    
    U = np.zeros((N, M+1))
    U[:,0] = np.sin(np.pi*x)
    
    r = dt/dx**2
    for j in range(0,M):
        for i in range(1, N-1):
            if i == 1:
                U[0,j] = 0
            elif i == N-1:
                U[N-1,j] = 0
            else:
                U[i,j+1] = U[i,j] + r*(U[i+1,j]-2*U[i,j]+U[i-1,j])
    return U


## GIVEN CONDITIONS FOR X AND T
# Interval for x creates the max and min values
xmin = 0
xmax = 1.0
#values of t
t = 0.0
tmax = 0.1

U = ode_solver(100, 100, xmin, xmax, t, tmax)
U = U[:,:-1]
x = np.linspace(xmin, xmax, 100)
t = np.linspace(t, tmax, 100)
T, X = np.meshgrid(t,x)
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X, T, U)
plt.title("Approximate Solution")
plt.show()


# exact solution
U_exact = np.zeros((100, 100))
for xj in x:
    xcount = 0
    for ti in t:
        tcount = 0
        U_exact[xcount, tcount] = math.exp(-(math.pi**2)*ti)*math.sin(math.pi*xj)
        tcount+=1
    xcount+=1
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X, T, U_exact)
plt.title("Exact Solution")
plt.show()
```
