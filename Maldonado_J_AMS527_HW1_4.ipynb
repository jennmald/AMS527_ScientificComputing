```python
import matplotlib.pyplot as plt

def g(x):
    return x+(x-1)**2

#def g(x):
    #return x**2 - 2

x_0 = 0.6
x_prev = g(x_0)
xlist = [x_0, x_prev]
x=0
xl=[]
yl=[]
while x < 1000:
    x_next = g(x_prev)
    x_prev = x_next
    xlist.append(x_prev)
    xl.append(x)
    yl.append(x)
    x+=1

plt.plot(xlist)
plt.title("Convergence of g(x)")
plt.show()
```
